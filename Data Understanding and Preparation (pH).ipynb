```python
# ==================
# Improts and Setup
# ==================

# Core libraries
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt

# Soil Theme Palette
PALETTE = {
    "clay_brown": "#8B5A2B",
    "deep_brown": "#5A3E2B",
    "green": "#2E7D32",
    "blue": "#1E88E5",
    "red": "#8B0000",
    "orange": "#F57C00",
    "yellow": "#FBC02D",
    "sand": "#F5E6C4",
    "charcoal": "#2B2B2B"
}

# Global style
plt.rcParams.update({
    "figure.figsize": (8, 5),
    "axes.facecolor": PALETTE["sand"],
    "figure.facecolor": PALETTE["sand"],
    "axes.edgecolor": PALETTE["charcoal"],
    "axes.labelcolor": PALETTE["charcoal"],
    "xtick.color": PALETTE["charcoal"],
    "ytick.color": PALETTE["charcoal"],
    "text.color": PALETTE["charcoal"],
    "axes.titleweight": "bold"
})
```

```python
# Update path as needed
df = pd.read_csv("wosis_latest_phaq.csv")

# Quick preview
df.head()
```

```python
# Structure
df.info()

# Summary stats
df.describe(include='all')

# Shape
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
```

```python
# Unique counts
print("Unique Profiles:", df['profile_id'].nunique())
print("Unique Layers:", df['layer_id'].nunique())
print("Regions:", df['region'].unique())
```

```python
# ===============
# Data Cleaning
# ==============

# Convert numeric fields
numeric_cols = ['upper_depth', 'lower_depth', 'value_avg', 'X', 'Y']

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Check missing values
df.isnull().sum()

# Drop rows missing critical fields
df_clean = df.dropna(subset=['upper_depth', 'lower_depth', 'X', 'Y']).copy()

print(f"Remaining rows after cleaning: {df_clean.shape[0]}")
```

```python
# =================
# Outlier Analysis
# =================

# Histogram
plt.hist(
    df_clean['value_avg'],
    bins=30,
    color=PALETTE["clay_brown"],
    edgecolor=PALETTE["deep_brown"]
)

plt.title("Distribution of pH Levels")
plt.xlabel("pH Level")
plt.ylabel("Frequency")

plt.show()

# Quick boxplot
plt.boxplot(
    df_clean['value_avg'],
    vert=False,
    patch_artist=True,
    boxprops=dict(facecolor=PALETTE["clay_brown"]),
    medianprops=dict(color=PALETTE["yellow"], linewidth=2)
)

plt.title("pH Spread")
plt.xlabel("pH Level")

plt.show()

# Optional: cap extreme values (adjust if needed)
lower_bound = df_clean['value_avg'].quantile(0.01)
upper_bound = df_clean['value_avg'].quantile(0.99)

df_clean = df_clean[
    (df_clean['value_avg'] >= lower_bound) & 
    (df_clean['value_avg'] <= upper_bound)
]
```

```python
# =====================
# Feature Engineering
# =====================

# Depth midpoint
df_clean['depth_mid'] = (df_clean['upper_depth'] + df_clean['lower_depth']) / 2

# Depth thickness
df_clean['depth_thickness'] = df_clean['lower_depth'] - df_clean['upper_depth']

# Categorize the depth
def categorize_depth(d):
    if d <= 20:
        return "Surface"
    elif d <= 50:
        return "Mid"
    else:
        return "Deep"

df_clean['depth_category'] = df_clean['depth_mid'].apply(categorize_depth)

# Rename coordinates
df_clean = df_clean.rename(columns={
    'X': 'longitude',
    'Y': 'latitude'
})
```

```python
# ========================
# Map and Sanity Check
# ========================

# Map Check
plt.figure(figsize=(8, 6))
plt.scatter(
    df_clean['longitude'],
    df_clean['latitude'],
    alpha=0.5
)
plt.title("Geographic Distribution of Soil Samples", color=PALETTE["charcoal"])
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()

# Sanity Check
# Depth consistency
invalid_depths = df_clean[df_clean['upper_depth'] > df_clean['lower_depth']]
print(f"Inconsistent depth rows: {len(invalid_depths)}")

# Remove invalid depths if needed
df_clean = df_clean[df_clean['upper_depth'] <= df_clean['lower_depth']]
```

```python
# Define and Classify pH Levels (Acid, Neutral, Base (Alkaline)

def classify_ph(pH):
    if pd.isna(pH):
        return None
    elif pH < 6.5:
        return "Acidic"
    elif pH <= 7.5:
        return "Neutral"
    else:
        return "Alkaline"

# Apply Classifer to data
df_clean['ph_category'] = df_clean['value_avg'].apply(classify_ph)
```
